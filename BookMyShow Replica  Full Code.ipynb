```python
class Movie():
    def __init__(self,row,column):
        self.row = row
        self.column = column
        self.booking = {}
        self.income = []
        
    def already_booked(self,row,column):
        seat_id = str(row)+str(column)
        if seat_id in self.booking:
            return True
        else:
            return False
        
    
    def show_seats(self):
        print("These are the available seats : ")
        for i in range(0,self.row+1): #Only by using 0 in the range gives us the row and column numbers.
            for j in range(0,self.column+1):
                if i == 0:
                    if j == 0:
                        print(" ",end = " ")
                    else:
                        print(j,end = " ")
                else:
                    if j == 0:
                        print(i,end = " ")
                    else:
                        if self.already_booked(i,j):
                            print("B",end = " ") #used to mark the already booked seats as "B" 
                        else:
                            print("S",end = " ")
                        
            print()
            
    def book_tickets(self):
        print("---Ticket Counter---")
        booking_row = int(input("Enter the row of the seat you want to book : "))
        booking_column = int(input("Enter the column of the seat you want to book : "))
        seat_id = str(booking_row)+str(booking_column)
        total_seats = self.row*self.column
        price = 0
        
        print("Ticket Pricing : ")
        if total_seats <= 60:
            price = 10
            print("The price of one ticket is $",price)
        if total_seats >= 60:
            half = self.row//2
            if booking_row <= half:
                price = 10
                print("The Price of one ticket is $",price)
            else:
                price = 8
                print("The price of one ticket is $",price)
        choice = int(input("Do you want to proceed your ticket booking \n1. Yes \n2. No"))
        if choice == 1:
            print("We are booking your ticket!!!")
            name = input("Enter your Name : ")
            gender = input("Enter your gender : ")
            age = input("Enter your age : ")
            mobile = input("Enter your mobile number : ")
            user_details = {'name':name,'gender':gender,'age':age,'mobile':mobile,'price':price}
            self.booking[seat_id]=user_details
            self.income.append(user_details['price'])
            print("Ticlet Booked Successfully")
        elif choice == 2:
            print("Booking cancelled!!!")
        
        
    def show_details(self,d_row,d_column):
        seat_id = str(d_row)+str(d_column)
        if seat_id in self.booking:
            user_details = self.booking[seat_id]
            print("Here are the booked user details : ")
            print("Nmae : " ,user_details['name'])
            print("Gender : " ,user_details['gender'])
            print("Age:" ,user_details['age'])
            print("Mobile Number :" ,user_details['mobile'])
            print("Tciket is booked for : $",user_details['price'])
        else:
            print("The Ticket is not yet booked!!!")
            
    def statistics(self,total_seat):
        current_income = 0
        self.total_seat = total_seat
        purchased_tickets = len(self.booking)
        percentage = round((purchased_tickets/self.total_seat)*100 , 2)
        total_income = 0
        first_half = (self.row//2)*self.column
        second_half = (self.total_seat-first_half)
        if total_seat <= 60:
            total_income = total_seat*10
        elif total_seat >= 60:
            total_income = (first_half*8)+(second_half*10)
        current_income = sum(self.income)
        print("The Statistics : ")
        print("The Number of purchased tickets : ",purchased_tickets)
        print("The Percentage of booked tickets : ",percentage)
        print("The Total income : ",total_income)
        print("The Current Income : ",current_income)
        
            
            
            
            
        
```

```python
row = int(input("Enter the number of rows : "))
column = int(input("Enter the number of columns : "))
m1 = Movie(row,column)
total_seat = row*column
while True:
    choice = int(input("\n1. Show the seats \n2. Buy a ticket \n3. Statistics \n4. Show the booked tickets user info \n0. Exit"))
    if choice == 1:
        print("Cinema:")
        m1.show_seats()
    elif choice == 2:
        m1.book_tickets()
    elif choice == 3:
        m1.statistics(total_seat)
    elif choice == 4:
        d_row = int(input("Enter the row number of the seat : "))
        d_column = int(input("Enter the column number of the seat : "))
        m1.show_details(d_row,d_column)
    elif choice == 0:
        print("You have opted to Exit!!!")
        break
    else:
        print("Enter a valid Choice : ")




            
```
