```python
class Food_Ordering_Application():
    
    def __init__(self,list_of_food):
        self.list_of_food = list_of_food
        
        self.user_data = {}
        self.list_of_user_data = []
        
        self.admin_data = {}
        self.list_of_admin_data = []
        
        self.new_user_data = {}
        self.new_list_of_user_data = []
        
        self.new_food_item = {}
        
        self.food_menu = {}
        self.ordered_list = []
   
    def user_sign_up(self):
        name = input("Enter the fullname of the user : ")
        email = input("Enter the email address of the user : ")
        mobile_number = int(input("Enter the mobile number of the user : "))
        address = input("Enter the user's address : ")
        password = input("Enter your password : ")
        self.user_data = {'Name':name,'email':email,'Mobile_number':mobile_number,'Address':address,'Password':password}
        self.list_of_user_data.append(self.user_data)
        
   
    def user_log_in(self):
        user_name = input("Enter your username : ")
        user_password = input("Enter your userpassword : ")
        if user_name == self.user_data['Name'] and user_password == self.user_data['Password']:
            print("Logged in Successfully!!!")
        else:
            print("User name or Password is Wrong!!!")
            
    def place_new_order(self):
        print("---Food Menu---")
        print("The following are the available food items in the menu : ")
        for i in self.list_of_food:
            print(f"{self.list_of_food.index(i)+1}. {i['Name']} ({i['Quantity']}) [{i['Price']}]")
        print("Order Your Food : ")
        for j in self.list_of_food:
            food_id = len(self.food_menu)+1
            self.food_menu[food_id]=j
        user_order = []
        n = int(input("Enter the number of items to order : "))
        user_order = list(map(int,input("\nEnter your order by giving the serial number displayed in the food menu : ").strip().split()))[:n]
        for k in user_order:
            self.ordered_list.append(self.food_menu[k])
        #print(self.ordered_list)
        
    def order_history(self):
        print("Your Order History : ")
        for l in self.ordered_list:
            print(l)
            
    def update_profile(self):
        print("Your Profile Details : ")
        for m in self.list_of_user_data:
            print(f"Name : {m['Name']} \nEmail : {m['email']} \nMobile Number : {m['Mobile_number']} \nAddress : {m['Address']} \nPassword : {m['Password']}")
        print("Please enter the desired changes you need to make : ")
        new_name = input("Enter the fullname of the user : ")
        new_email = input("Enter the email address of the user : ")
        new_mobile_number = int(input("Enter the mobile number of the user : "))
        new_address = input("Enter the user's address : ")
        new_password = input("Enter your password : ")
        self.new_user_data = {'New_Name':new_name,'New_Email':new_email,'New_Mobile_number':new_mobile_number,'New_Address':new_address,'New_Password':new_password}
        self.new_list_of_user_data.append(self.new_user_data)
        print("Your Updated Profile Details : ")
        for n in self.new_list_of_user_data:
            print(f"Name : {n['New_Name']} \nEmail : {n['New_Email']} \nMobile Number : {n['New_Mobile_number']} \nAddress : {n['New_Address']} \nPassword : {n['New_Password']}")

        
    def admin_sign_up(self):
        name = input("Enter the fullname of the admin : ")
        email = input("Enter the email address of the admin : ")
        mobile_number = int(input("Enter the mobile number of the admin : "))
        address = input("Enter the admin's address : ")
        password = input("Enter your password : ")
        self.admin_data = {'Name':name,'email':email,'Mobile_number':mobile_number,'Address':address,'Password':password}
        self.list_of_admin_data.append(self.admin_data)
    
    
    def admin_log_in(self):
        admin_name = input("Enter your username : ")
        admin_password = input("Enter your password : ")
        if admin_name == self.admin_data['Name'] and admin_password == self.admin_data['Password']:
            print("Logged in Successfully!!!")
        else:
            print("User name or Password is Wrong!!!")  
            
    def add_food_items(self):
        num = 5
        for o in range(num):
            keys = input() 
            values = (input()) 
            self.new_food_item[keys] = values
        self.list_of_food.append(self.new_food_item)
        for p in self.list_of_food:
            food_id = len(self.food_menu)+1
            self.food_menu[food_id]=p
        print(self.list_of_food)
        
    def edit_food_items(self):
        print("---Food Menu---")
        print("The following are the available food items in the menu : ")
        for q in self.list_of_food:
            print(f"{self.list_of_food.index(q)+1}. {q['Name']} ({q['Quantity']}) [{q['Price']}]")
        for r in self.list_of_food:
            food_id = len(self.food_menu)+1
            self.food_menu[food_id]=r
        edit = int(input("Enter the Food ID of the item you want to edit : "))
        edit_item = input("Enter the field you desire to edit : ")
        new_item = input("Enter the new item : ")
        self.food_menu[edit][edit_item] = new_item
        print("The Updated Food Menu : ")
        for s in self.list_of_food:
            print(f"{self.list_of_food.index(s)+1}. {s['Name']} ({s['Quantity']}) [{s['Price']}]")
        
    def remove_food_items(self):
        print("---Food Menu---")
        print("The following are the available food items in the menu : ")
        for t in self.list_of_food:
            print(f"{self.list_of_food.index(t)+1}. {t['Name']} ({t['Quantity']}) [{t['Price']}]")
        for u in self.list_of_food:
            food_id = len(self.food_menu)+1
            self.food_menu[food_id]=u
        remove = int(input("Enter the food id of the item you want to remove : "))
        del self.food_menu[remove]
        print("Food Item Removed Successfully!!!")
    
        
    def view_food_items(self):
        print("These are the available Food Items: ")
        for v in self.list_of_food:
            print(f"{self.list_of_food.index(v)+1}. {v['Name']} ({v['Quantity']}) [{v['Price']}]")
        #print(self.food_menu)
        
  
                
    
    
    
                  

        
        
```

```python
a = Food_Ordering_Application(list_of_food = [{"Name":"Tandoori Chicken","Quantity":"4 Pieces","Price":"INR 240"},{"Name":"Vegan Burger","Quantity":"1 Piece","Price":"INR 320"},{"Name":"Truffle Cake","Quantity":"500mgs","Price":"INR 900"}])
while True:
    choice = int(input("Enter your Choice : \n1. User Sign Up \n2. User Log In \n3. Admin Sign Up \n4. Admin Log In \n0. Exit "))
    if choice == 1:
        a.user_sign_up()
    elif choice == 2:
        a.user_log_in()
        preference = int(input("Enter your preference : \n1. Place new order \n2. Order History \n3. Update Profile"))
        if preference == 1:
            a.place_new_order()
        elif preference == 2:
            a.order_history()
        elif preference == 3:
            a.update_profile()
    elif choice == 3:
        a.admin_sign_up()
    elif choice == 4:
        a.admin_log_in()
        need = int(input("What editing do you want to do : \n1.Add Food Items \n2. Edit Food Items \n3. View all Food Items \n4. Remove Food Items"))
        if need == 1:
            a.add_food_items()
        elif need == 2:
            a.edit_food_items()
        elif need == 3:
            a.view_food_items()
        elif need == 4:
            a.remove_food_items()
        else:
            print("Enter a valid option")
    elif choice == 0:
        print("You have opted to exit!!!")
        break
    else:
        print("Enter a valid choice")
```
